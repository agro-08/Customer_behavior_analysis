```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.median())
df["Review Rating"]
```

```python
df = df.rename(columns={
    '_c_u_s_t_o_m_e_r___i_d_': 'customer_id',
    '_a_g_e_': 'age',
    '_g_e_n_d_e_r_': 'gender',
    '_i_t_e_m___p_u_r_c_h_a_s_e_d_': 'item_purchased',
    '_c_a_t_e_g_o_r_y_': 'category',
    '_p_u_r_c_h_a_s_e___a_m_o_u_n_t___(_u_s_d_)_': 'purchase_amount',
    '_l_o_c_a_t_i_o_n_': 'location',
    '_s_i_z_e_': 'size',
    '_c_o_l_o_r_': 'color',
    '_s_e_a_s_o_n_': 'season',
    '_r_e_v_i_e_w___r_a_t_i_n_g_': 'review_rating',
    '_s_u_b_s_c_r_i_p_t_i_o_n___s_t_a_t_u_s_': 'subscription_status',
    '_s_h_i_p_p_i_n_g___t_y_p_e_': 'shipping_type',
    '_d_i_s_c_o_u_n_t___a_p_p_l_i_e_d_': 'discount_applied',
    '_p_r_o_m_o___c_o_d_e___u_s_e_d_': 'promo_code_used',
    '_p_r_e_v_i_o_u_s___p_u_r_c_h_a_s_e_s_': 'previous_purchases',
    '_p_a_y_m_e_n_t___m_e_t_h_o_d_': 'payment_method',
    '_f_r_e_q_u_e_n_c_y___o_f___p_u_r_c_h_a_s_e_s_': 'frequency_of_purchases'
})
df
```

```python
#create new column named age_group
labels=["Young Adult" , "Adult" , "Middle-aged" , "Senior" ]
df["age_group"]=pd.qcut(df["age"],q=4 , labels = labels)
```

```python
#create column purchase_frequency_days 
frequency_purchase={
    "Fortnightly":14 ,
    "Weekly":7 ,
    "Annually":365 , 
    "Quarterly":90 , 
    "Bi-Weekly":14 , 
    "Monthly":30 , 
    "Every 3 Months":90
}
df["frequency_purchase_days"]=df["frequency_of_purchases"].map(frequency_purchase)
df[["frequency_purchase_days" ,"frequency_of_purchases"]].head(10)
```

```python
df = df.drop("promo_code_used" , axis=1)
```
